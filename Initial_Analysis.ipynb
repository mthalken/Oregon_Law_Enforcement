```python
crimes = crimes.loc[crimes['state_abbr'] == 'OR'].reset_index() 
crimes
```

```python
crimes2 = crimes.drop(['index', 'state_abbr', 'state_name', 'state_name', 'caveats'], axis=1)

rape_columns = crimes2.iloc[:, 4:6]
rape_legacy = rape_columns.iloc[:, 0].dropna()
rape_revised = rape_columns.iloc[:, 1].dropna()
rape_revised
```

```python
rape_legacy2 = rape_legacy.iloc[:-4]
```

```python
type(rape_revised)
```

```python
rape = pd.DataFrame(rape_legacy2.append(rape_revised))
rape.rename(columns={0:'rape'})
```

```python
crimes3 = crimes2.join(rape)
crimes4 = crimes3.drop(['rape_legacy', 'rape_revised'], axis=1)
crimes4
```

```python
stats = crimes4.iloc[:, :2]
stats
```

```python
crime_df = crimes4.drop(['year', 'population'], axis=1)
crime_sum = crime_df.sum(axis=1)
crime_sum_df = pd.DataFrame(crime_sum, columns=['sum'])
crime_sum_df
```

```python
crime_final = stats.join(crime_sum_df)
crime_final
```

```python
pop = crime_final['population']
pop
```

```python
year = crime_final['year']
year
```

```python
plt.plot(year, pop)
plt.show()
```
